```python
%matplotlib inline
import numpy as np
from scipy import fftpack

# Assume `fish_counts` is a list of the number of fish you caught each day
# for a full year, starting on January 1st and ending on December 31st.
# We'll just generate some random data for this example.
np.random.seed(0)  # for reproducibility
fish_counts = np.random.poisson(lam=20, size=365)

# Perform a Fourier transform on the data to convert it to the frequency domain.
fish_fft = fftpack.fft(fish_counts)

# The frequencies associated with each Fourier coefficient are given by the FFT frequencies.
frequencies = fftpack.fftfreq(len(fish_counts))

# Now, `fish_fft` contains the Fourier coefficients for each frequency in `frequencies`.
# The absolute value of each coefficient tells us the amplitude of that frequency component,
# and the angle of each coefficient (in polar coordinates) tells us the phase.

# We can plot the spectrum to see which frequencies are most prominent.
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(frequencies, np.abs(fish_fft))
plt.xlabel('Frequency (cycles per day)')
plt.ylabel('Amplitude')
plt.title('Spectrum of fish counts')
plt.show()
```
